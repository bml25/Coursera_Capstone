### Import Libraries

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from urllib.request import urlopen
import ssl
import csv
```

### Set the Parameters for ssl

```python
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE
```

### Establish a Source to Import Data From

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

soup = BeautifulSoup(source, 'lxml')
```

### Locate the Table on the Website

```python
table = soup.find('table',{'class':'wikitable sortable'})
```

### Define the Table Rows

```python
table_rows = table.find_all('tr')
```

### Transform Website Data into a Pandas Dataframe

```python
data = []
for row in table_rows:
    data.append([t.text.strip() for t in row.find_all('td')])

df = pd.DataFrame(data, columns = ['PostalCode', 'Borough', 'Neighbourhood'])

df
```

### Delete Rows where the PostalCode Column is Blank

```python
df = df[~df['PostalCode'].isnull()]

df
```

### Drop Rows where the Borough is Not Assigned

```python
df.drop(df[df['Borough']=="Not assigned"].index, axis = 0, inplace = True)

df
```

### Replace Blank Neighbourhood Values with the Bourough Value

```python
df.loc[df['Neighbourhood']=="Not assigned",'Neighbourhood']=df.loc[df['Neighbourhood']=="Not assigned",'Borough']

df
```

### Reset the Dataframe Index

```python
df.reset_index(drop = True, inplace = True)

df
```
